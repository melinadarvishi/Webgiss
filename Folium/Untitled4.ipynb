```python
import folium

# ایجاد نقشه اصلی
m = folium.Map(location=[36.4706, 52.3466], zoom_start=12)

# ۱. لایه خیابان (پیش‌فرض)
folium.TileLayer('openstreetmap', name='Street Map').add_to(m)

# ۲. لایه ماهواره‌ای گوگل (بسیار دقیق برای نقشه‌برداری)
folium.TileLayer(
    tiles='https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}',
    attr='Google',
    name='Google Satellite'
).add_to(m)

# اضافه کردن منوی لایه‌ها
folium.LayerControl().add_to(m)

# نمایش نهایی
m

m.save("map4_layers.html")
```
